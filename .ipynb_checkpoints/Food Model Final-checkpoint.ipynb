```python
import pandas as pd
import numpy as np
train=pd.read_csv("/Users/Thiru/Desktop/Pricing & Revenue Mgmt./Food_Delivery_Project/Data/train_full.csv")
test=pd.read_csv("/Users/Thiru/Desktop/Pricing & Revenue Mgmt./Food_Delivery_Project/Data/test_full.csv")
train.columns
```

```python
train_grouped=train.groupby('week').agg({'checkout_price' : np.sum,
                           'num_orders' : np.sum,
                          'homepage_featured' : np.sum})
train_grouped.reset_index()
train_grouped
```

```python
import statsmodels.api as sm
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import mean_squared_error
import math

X = train_grouped[['checkout_price','homepage_featured']]
y = train_grouped["num_orders"]

X_train = X.iloc[0:100:,]
X_test = X.iloc[100:,]
y_train = y.iloc[0:100:]
y_test = y.iloc[100:]

# Note the difference in argument order
model = sm.OLS(y_train, X_train).fit()
predictions = model.predict(X_test) # make the predictions by the model
from sklearn.metrics import mean_squared_error
print(math.sqrt(mean_squared_error(y_test, predictions)))

# Print out the statistics
model.summary()
```

```python
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif.round(1)
```

```python
import matplotlib.pyplot as plt
#plt.plot(y_train)
plt.plot(y_test)
plt.plot(predictions)
#plt.plot(X_train)
plt.show()
```

```python
Residuals=pd.Series(y_test - predictions)
plt.plot(Residuals)
```

```python
from statsmodels.tsa.stattools import adfuller
series = Residuals.diff().dropna()
plt.plot(series)
X=series.values
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
```
